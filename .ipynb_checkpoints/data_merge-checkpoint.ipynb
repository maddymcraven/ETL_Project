```python
import pandas as pd
```

```python
#load file
file_path = 'usa_county_wise.csv'
covid_data = pd.read_csv(file_path)
covid_data
```

```python
#clean data
filter = (covid_data.iso2 == "US")
c1 = covid_data[filter]
c2 = c1[['FIPS', 'Admin2', 'Province_State', 'Lat', 'Long_', 'Date', 'Confirmed', 'Deaths']]
c3 = c2.rename(columns ={
    "Admin2":"city",
    "Province_State":"state",
    "Long_":"lng",
    "Lat":'lat'
})
filter =  (c3.Date == "7/27/20")
c4 = c3[filter]
covid_clean = c4.reset_index()
covid_clean
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
covid_clean.to_sql('covid_data', con=engine)
sql_data = engine.execute("SELECT * FROM covid_data").fetchall()
sql_data
```
